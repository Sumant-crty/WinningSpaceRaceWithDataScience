```python
# Create a new map centered on the mean coordinates (using nasa_coordinate as a general starting point)
launch_outcomes_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Iterate through each launch record in spacex_df
for index, row in spacex_df.iterrows():
    # Determine marker color based on 'class' column
    # Green for successful launches (class = 1)
    # Red for unsuccessful launches (class = 0)
    marker_color = 'green' if row['class'] == 1 else 'red'

    # Create a CircleMarker for each launch
    folium.CircleMarker(
        location=[row['Lat'], row['Long']],
        radius=5, # Small radius for individual launches
        color=marker_color,
        fill=True,
        fill_color=marker_color,
        fill_opacity=0.7,
        popup=f"Launch Site: {row['Launch Site']}<br>Outcome: {'Success' if row['class'] == 1 else 'Failure'}"
    ).add_to(launch_outcomes_map)

# Display the map
launch_outcomes_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

### Calculate the distance from a launch site to a coastline

Let's choose the first launch site, 'CCAFS LC-40', from `launch_sites_df` and a sample coastline point to calculate the distance. We will then visualize this on the `site_map`.

```python
# Select a launch site (e.g., the first one in launch_sites_df)
launch_site = launch_sites_df.loc[0]
launch_site_lat = launch_site['Lat']
launch_site_lon = launch_site['Long']

# Sample coastline coordinate (from the example in the prompt)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate the distance to the coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance from {launch_site['Launch Site']} to coastline: {distance_coastline:.2f} km")

# Add a marker for the coastline point
folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#0000FF;"><b>Coastline</b></div>',
    )
).add_to(site_map)

# Draw a line between the launch site and the coastline
points = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
folium.PolyLine(points, color='blue', weight=2.5, opacity=1).add_to(site_map)

# Display the map with the new additions
site_map
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property
# for example
distance_marker = folium.Marker(
   [coastline_lat, coastline_lon],
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline),
       )
   )
site_map.add_child(distance_marker)
site_map
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
lines = folium.PolyLine(locations=points, weight=1)
site_map.add_child(lines)
```

### Calculate distances to other proximities (city, railway, highway)

```python
# Sample coordinates for a city, railway, and highway near CCAFS LC-40
# These are illustrative coordinates and might not be exact real-world locations
city_lat, city_lon = 28.61200, -80.80770 # Example: Titusville, FL
railway_lat, railway_lon = 28.56300, -80.57560 # Example: Near Cape Canaveral rail line
highway_lat, highway_lon = 28.56270, -80.57060 # Example: Near FL-401

# Calculate distances
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

print(f"Distance from {launch_site['Launch Site']} to City: {distance_city:.2f} km")
print(f"Distance from {launch_site['Launch Site']} to Railway: {distance_railway:.2f} km")
print(f"Distance from {launch_site['Launch Site']} to Highway: {distance_highway:.2f} km")

# Function to add a marker with distance and draw a line
def add_distance_marker_and_line(map_obj, start_lat, start_lon, end_lat, end_lon, distance, label, line_color='green', marker_color='#d35400'):
    # Add marker for the end point
    folium.Marker(
        [end_lat, end_lon],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12; color:{marker_color};"><b>{label}: {distance:.2f} KM</b></div>',
        )
    ).add_to(map_obj)

    # Draw a line between the launch site and the end point
    points = [[start_lat, start_lon], [end_lat, end_lon]]
    folium.PolyLine(points, color=line_color, weight=2.5, opacity=1).add_to(map_obj)

# Add markers and lines for City, Railway, and Highway
add_distance_marker_and_line(site_map, launch_site_lat, launch_site_lon, city_lat, city_lon, distance_city, 'City', line_color='purple')
add_distance_marker_and_line(site_map, launch_site_lat, launch_site_lon, railway_lat, railway_lon, distance_railway, 'Railway', line_color='orange')
add_distance_marker_and_line(site_map, launch_site_lat, launch_site_lon, highway_lat, highway_lon, distance_highway, 'Highway', line_color='red')

# Display the updated map
site_map
```
